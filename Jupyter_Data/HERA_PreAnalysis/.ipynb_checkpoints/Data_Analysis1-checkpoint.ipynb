```python
# import all necessary packages and functions.
import csv
from datetime import datetime, timedelta as td
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
import seaborn as sns
%matplotlib inline
```

```python
# Display the first few rows of the HR data file. df reads as dataframe
df_HR = pd.read_csv("HR.csv")
display(df_HR.head())
df_HR.info()

df_BP = pd.read_csv("BP.csv")
display(df_BP.head())
df_BP.info()
```

```python
# Taking an Astroskin dataframe relative to the biometrics being studied and convert it to a good time representation
# With consideration of the Timezone where the data as been recorded
def time_transform(data, hours, unit):    
    display(data.head())
    
    # Renaming columns to meaningful names
    data.columns = ['TIME', unit]
    
    # Removing the 256Hz frequency embedded
    data['TIME'] = data['TIME']/256
    
    # Conversion of the epoch Timestamps with seconds as unit specified - Output object is datetime64[ns]
    data['TIME'] = pd.to_datetime(data['TIME'],unit='s')
    display(data.head())
    
    # Checking if the UTC conversion is positive or negative
    if hours < 0:
        data['TIME'] = data['TIME'] - pd.offsets.Hour(abs(hours)).delta
    elif hours > 0:
        data['TIME'] = data['TIME'] + pd.offsets.Hour(abs(hours)).delta
    display(data.head())
    
    # Removing ms and ns for 1Hz incrementation
    data['TIME'] = data['TIME'].values.astype('<M8[s]')
    display(data.head())
    
    # Creating a new column Date by stripping the Date from the overall datatime object
    # Type Coercion from strftime method resulting as a new time object represented as a str for both Time and Date columns
    data['DATE']= data['TIME']
    data['DATE'] = data['DATE'].apply(lambda x:x.date().strftime('%y-%m-%d'))
    data['TIME'] = data['TIME'].apply(lambda x:x.time().strftime('%H:%M:%S'))
    data = data[['DATE', 'TIME', unit]]

    # Displaying head and data types for final result
    display(data.head())
    data.dtypes
    
    return data
```

```python
df_HR = time_transform(df_HR, -8, 'BPM')
```

```python
df_BP = time_transform(df_BP, -8, 'mmHg')
```

```python
def filter_data(data, condition):
    """
    Remove elements that do not match the condition provided.
    Takes a dataframe column as input and returns a filtered list.
    Conditions should be a list of strings of the following format:
      '<field> <op> <value>'
    where the following operations are valid: >, <, >=, <=, ==, !=
    
    Example: ["duration < 15", "start_city == 'San Francisco'"]
    """

    # Only want to split on first two spaces separating field from operator and
    # operator from value: spaces within value should be retained.
    field, op, value = condition.split(" ", 2)
    print('Chosen field for filter: ' + field)
    # check if field is valid
    if field not in data.columns.values :
        raise Exception("'{}' is not a feature of the dataframe. Did you spell something wrong?".format(field))

    # convert value into number or strip excess quotes if string
    try:
        value = float(value)
    except:
        value = value.strip("\'\"")

    # get booleans for filtering
    if op == ">":
        matches = data[field] > value
    elif op == "<":
        matches = data[field] < value
    elif op == ">=":
        matches = data[field] >= value
    elif op == "<=":
        matches = data[field] <= value
    elif op == "==":
        matches = data[field] == value
    elif op == "!=":
        matches = data[field] != value
    else: # catch invalid operation codes
        raise Exception("Invalid comparison operator. Only >, <, >=, <=, ==, != allowed.")
    
    # filter data and outcomes
    data = data[matches].reset_index(drop = True)
    display(data.head())
    return data
```

```python
df_BP = filter_data(df_BP, "mmHg > 0")
```

```python
df_BP.plot(x = 'TIME',y = 'mmHg', title = 'mmHg in Time', color = 'red')
```
